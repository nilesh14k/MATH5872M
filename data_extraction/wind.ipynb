```python
import os
import pandas as pd
from tqdm import tqdm

# Directory containing the .lst files
wind_raw_dir = 'D:\\_NILESH\\dissertation\\data\\raw\\wind\\'
wind_temp='D:\\_NILESH\\dissertation\\data\\raw\\wind_temp\\'

# Output directory for the CSV files
data_dir = 'D:\\_NILESH\\dissertation\\data\\'

# Column names according to the specified format
columns = [
    "Year", "Day", "Hour", "Minute",
    "Field magnitude average, nT", "RMS SD B scalar, nT",
    "Speed, km/s", "Proton Density, n/cc",
    "Proton Temperature, K", "Flow pressure, nPa",
    "Electric field, mV/m"
]

# Get the list of .lst files
lst_files = [filename for filename in os.listdir(wind_raw_dir) if filename.endswith(".lst")]

# Iterate through the files in the directory with a progress bar
for filename in tqdm(lst_files, desc='Processing files', unit='file'):
    filepath = os.path.join(wind_raw_dir, filename)
    
    # Create a list to hold the rows
    data_rows = []

    with open(filepath, 'r') as file:
        # Read each line from the file
        for line in file:
            # Split the line by whitespace and convert the values to the appropriate types
            values = line.split()
            if len(values) < 11:
                continue
            values = [int(values[0]), int(values[1]), int(values[2]), int(values[3])] + [float(v) for v in values[4:]]
            data_rows.append(values)

    # Create a DataFrame from the rows
    data_df = pd.DataFrame(data_rows, columns=columns)

    # Save the DataFrame to a CSV file in the specified data directory, using the same name as the original .lst file
    csv_filename = filename.replace('.lst', '.csv')
    data_df.to_csv(os.path.join(wind_temp, csv_filename), index=False)
```

```python
# Output file for the combined CSV
output_file = 'D:\\_NILESH\\dissertation\\data\\solar_wind.csv'

# Get the list of CSV files
csv_files = [filename for filename in os.listdir(wind_temp) if filename.endswith(".csv")]

# Initialize an empty DataFrame to hold the combined data
combined_data_df = pd.DataFrame()

# Iterate through the CSV files in the directory with a progress bar
for filename in tqdm(csv_files, desc='Combining files', unit='file'):
    filepath = os.path.join(wind_temp, filename)
    # Read the CSV file into a DataFrame
    data_df = pd.read_csv(filepath)
    # Append the DataFrame to the combined data
    combined_data_df = combined_data_df.append(data_df, ignore_index=True)

# Save the combined DataFrame to the output CSV file
combined_data_df.to_csv(output_file, index=False)

print(f"All CSV files have been combined into {output_file}")
```
