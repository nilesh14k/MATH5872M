```python
import pandas as pd
import matplotlib.pyplot as plt

# Change the default text color
plt.rcParams['text.color'] = 'black'
plt.rcParams['axes.labelcolor'] = 'white'
plt.rcParams['xtick.color'] = 'white'
plt.rcParams['ytick.color'] = 'white'

data_dir='D:\\_NILESH\\dissertation\\data\\'
df_wind=pd.read_csv(data_dir+'solar_wind.csv')
df_sun=pd.read_csv(data_dir+'sunspots_temp.csv')
df_flare=pd.read_csv(data_dir+'solar_flare.csv')
df_flare['Date']=df_flare['start.date']
df_cme=pd.read_csv(data_dir+'cme.csv')

# Sample the sunspot numbers for months
df_sun["Date"] = pd.to_datetime(df_sun["Date"])
df_sun.set_index("Date", inplace=True)
monthly_sunspot_mean = df_sun.resample("Q")["Monthly mean total sunspot number"].mean()

# Sample the solar flare counts for years
df_flare["Date"] = pd.to_datetime(df_flare["Date"])
df_flare.set_index("Date", inplace=True)
yearly_flare_count = df_flare.resample("Q").size()

# Sample the CME counts for months
df_cme["Date"] = pd.to_datetime(df_cme["Date"])
df_cme.set_index("Date", inplace=True)
monthly_cme_count = df_cme.resample("Q").size()

# Combine the sampled data into a single dataframe
df_combined = pd.DataFrame({
    "Sunspot_Number": monthly_sunspot_mean,
    "Solar_Flare_Count": yearly_flare_count,
    "Corona_Mass_Ejection_Count": monthly_cme_count
})

# Plotting the time series graph
plt.figure(figsize=(12, 6))

# Plot sunspot numbers
plt.plot(df_combined.index, df_combined["Sunspot_Number"], label="Sunspot", color="red")

# Plot solar flare counts
plt.plot(df_combined.index, df_combined["Solar_Flare_Count"], label="Solar Flare", color="darkgrey")

# Plot CME counts
plt.plot(df_combined.index, df_combined["Corona_Mass_Ejection_Count"], label="CME", color="green")

# Add vertical lines at every 11-year interval
start_year = 1986
end_year = df_combined.index.year[-1]
for year in range(start_year, end_year + 1, 11):
    plt.axvline(pd.Timestamp(year, 1, 1), color="lightgrey", linestyle="--", alpha=0.5)

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Count / Sunspot Number")
plt.title("Time Series Global Graph")
plt.legend()

# Save the plot with a transparent background
plt.savefig('timeseries_plot.png', transparent=True)

# Show the plot
plt.show()
```

```python
def parse_energy_range(energy_range):
    try:
        low, high = map(float, energy_range.split('-'))
        return (low + high) / 2
    except:
        return None

df_flare['average_energy'] = df_flare['energy.kev'].map(parse_energy_range)
df_flare['Energy_Calculated'] = df_flare['average_energy'] * df_flare['total.counts']
```

```python
# Convert 'Kinetic_Energy' column to numeric, forcing non-numeric values to NaN
df_cme['Kinetic_Energy'] = pd.to_numeric(df_cme['Kinetic_Energy'], errors='coerce')

# Converting CME energy to keV
df_cme['Kinetic_Energy_keV'] = df_cme['Kinetic_Energy'] / 1.0e+3


# Sample the energy data for CME and Solar flares for months
monthly_cme_energy = df_cme.resample("M")["Kinetic_Energy_keV"].mean()

monthly_flare_energy = df_flare.resample("M")["Energy_Calculated"].mean()

# Combine the sampled data into a single dataframe
df_combined_energy = pd.DataFrame({
    "Sunspot_Number": monthly_sunspot_mean,
    "Solar_Flare_Energy": monthly_flare_energy,
    "Corona_Mass_Ejection_Energy": monthly_cme_energy,
})


# # Plotting
# plt.figure(figsize=(12, 8))

# plt.plot(monthly_cme_energy.index, monthly_cme_energy, label='CME Energy (keV)')
# plt.plot(monthly_flare_energy.index, monthly_flare_energy, label='Solar Flare Energy (keV)')
# plt.plot(monthly_sunspot_mean.index, monthly_sunspot_mean, label='Sunspot Number') # Updated here

# plt.xlabel('Time')
# plt.ylabel('Energy (keV) / Sunspot Number')
# plt.title('Energy Comparison of CME and Solar Flares with Sunspot Number Over Time')
# plt.legend()

# # Saving the plot
# plt.savefig('energy_comparison.png')
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Define the start year
start_year = 1986

# Define the end year
end_year = df_sun.index.year.max()

# Calculate the number of cycles
n_cycles = int((end_year - start_year) / 11)

# Generate a list of years where each cycle starts
cycle_years = [start_year + i*11 for i in range(n_cycles + 1)]

# Plotting
plt.figure(figsize=(12, 8))

plt.plot(df_sun.index, df_sun['Monthly mean total sunspot number'], color='white')

plt.xlabel('Time', color='white')
plt.ylabel('Sunspot Number', color='white')
plt.title('Sunspot Number Over Time', color='white')

# Add vertical lines for each solar cycle
for year in cycle_years:
    plt.axvline(pd.to_datetime(str(year)), color='white', linestyle='dashed')

plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')

# plt.grid(True)

# Saving the plot with transparency
plt.savefig('sunspot_number_with_cycles.png', transparent=True)
```
