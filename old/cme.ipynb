```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
# data = pd.read_csv("your_data.csv")

# Convert the Date column to a datetime format
data["Date"] = pd.to_datetime(data["Date"])

# Clean the "Kinetic Energy" column and convert it to numeric format
data["Kinetic_Energy"] = pd.to_numeric(data["Kinetic_Energy"], errors="coerce")

# Set the Date column as the index
data.set_index("Date", inplace=True)

# Resample the data to monthly intervals and calculate the average of Kinetic Energy
monthly_avg = data["Kinetic_Energy"].resample("M").mean()

# Plotting the time series graph for average Kinetic Energy over monthly intervals
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg.index, monthly_avg, color="blue")

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Average Kinetic Energy")
plt.title("Average Kinetic Energy over Monthly Intervals (1996 to 2008)")

# Show the plot
plt.show()
```

```python
data["Mass"] = pd.to_numeric(data["Mass"], errors="coerce")

# Resample the data to monthly intervals and calculate the average of Mass
monthly_avg_mass = data["Mass"].resample("M").mean()

# Plotting the time series graph for average Mass over monthly intervals
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg_mass.index, monthly_avg_mass, color="green")

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Average Mass")
plt.title("Average Mass over Monthly Intervals")

# Show the plot
plt.show()
```

```python
data["Accel"] = pd.to_numeric(data["Accel"], errors="coerce")

monthly_avg_accel = data["Accel"].resample("M").mean()

# Plotting the time series graph for average Mass over monthly intervals
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg_accel.index, monthly_avg_accel, color="green")

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Average Accel")
plt.title("Average Accel over Monthly Intervals")

# Show the plot
plt.show()
```
