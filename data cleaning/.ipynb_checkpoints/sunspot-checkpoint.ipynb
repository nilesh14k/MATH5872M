```python
import warnings
warnings.filterwarnings("ignore")
import logging
logging.basicConfig(level=logging.ERROR)
import pandas as pd
data_dir='D:\\_NILESH\\dissertation\\data\\sunspots.csv'
```

```python
data = pd.read_csv(data_dir)
data.head()
```

```python
data.dtypes
```

```python
data['date'] = pd.to_datetime(data['Year'].astype(int).astype(str) + '-' + 
                               data['Month'].astype(int).astype(str) + '-' + 
                               data['Day'].astype(int).astype(str), format='%Y-%m-%d')

# Filter the data to include only dates on or after 2000-01-01
data = data[data['date'] >= '2000-01-01']

# Reset the index
data.reset_index(drop=True, inplace=True)

# Select the required columns and rename to lowercase
new_data = data[['date', 'Corrected whole spot area', 'Carrington Longitude', 'Latitude', 'Number of spots in group', 'Zurich/McIntosh group type', 'Magnetic group type']].copy()
new_data.columns = ['date', 'area', 'longitude', 'latitude', 'number_of_spots_in_group', 'zurich/mcintosh_group_type', 'magnetic_group_type']

print(new_data.size)
new_data.head()
```

```python
new_data.dropna(inplace=True)
print(new_data.size)
new_data.head()
```

```python
new_data['number_of_spots_in_group'] = new_data['number_of_spots_in_group'].replace(0.0, 1.0)
new_data['area'] = new_data['area'].replace(0.000e+00, 1.000e+00)
# Remove rows with missing values (i.e., rows with '  ') in the 'magnetic_group_type' column
new_data = new_data[new_data['magnetic_group_type'] != '   ']
new_data.size
```

```python
print(new_data['area'].unique())
new_data['area'].value_counts().unique()
```

**1. Introduction**

While both these phenomena have been studied extensively, the intricate relationship between them remains a subject of keen interest. This chapter delves into the association between sunspot characteristics and the intensity of CMEs, offering insights into the solar dance of these two phenomena.
