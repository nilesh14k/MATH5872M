```python
import warnings
warnings.filterwarnings("ignore")
import logging
logging.basicConfig(level=logging.ERROR)
import pandas as pd

data_dir='D:\\_NILESH\\dissertation\\data\\raw\\flare\\'
```

```python
df1 = pd.read_csv(data_dir+'ace_epam_5m_data.csv')
df1.head()
```

```python
def compute_energy_released(row, flux_column, energy_range):
    # Convert the flux value from MeV to keV
    flux_kev = row[flux_column] * 1000
    # Extract the energy range values
    energy_start, energy_end = map(int, energy_range.split('_'))
    # Compute the average energy
    avg_energy = (energy_start + energy_end) / 2
    # Return the computed energy released
    return flux_kev * avg_energy

# Define the flux columns and their corresponding energy ranges
flux_columns = {
    'electron_flux_38_53 (particles/cm2-s-ster-MeV)': '38_53',
    'electron_flux_175_315 (particles/cm2-s-ster-MeV)': '175_315',
    'proton_flux_47_68 (particles/cm2-s-ster-MeV)': '47_68',
    'proton_flux_115_195 (particles/cm2-s-ster-MeV)': '115_195',
    'proton_flux_310_580 (particles/cm2-s-ster-MeV)': '310_580',
    'proton_flux_795_1193 (particles/cm2-s-ster-MeV)': '795_1193',
    'proton_flux_1060_1900 (particles/cm2-s-ster-MeV)': '1060_1900'
}

# For each flux column, compute the energy released and store in a new column
for flux_column, energy_range in flux_columns.items():
    energy_column = flux_column.replace('flux', 'energy_released').replace('MeV', 'keV').replace('cm2', 'km2')
    df1[energy_column] = df1.apply(compute_energy_released, flux_column=flux_column, energy_range=energy_range, axis=1)
    
    # Convert the values from cm² to km²
    df1[energy_column] /= 1e10

# Display the updated dataframe
df1.head()
```

```python
# Define the energy released columns for electrons and protons
electron_energy_columns = [
    'electron_energy_released_38_53 (particles/km2-s-ster-keV)',
    'electron_energy_released_175_315 (particles/km2-s-ster-keV)'
]

proton_energy_columns = [
    'proton_energy_released_47_68 (particles/km2-s-ster-keV)',
    'proton_energy_released_115_195 (particles/km2-s-ster-keV)',
    'proton_energy_released_310_580 (particles/km2-s-ster-keV)',
    'proton_energy_released_795_1193 (particles/km2-s-ster-keV)',
    'proton_energy_released_1060_1900 (particles/km2-s-ster-keV)'
]

# Calculate the total energy released for electrons and protons
df1['total_electron_energy_released (keV)'] = df1[electron_energy_columns].sum(axis=1)
df1['total_proton_energy_released (keV)'] = df1[proton_energy_columns].sum(axis=1)

# Display the updated dataframe with new columns
df1.head()
```

```python
selected_df1 = df1[['time (yyyy MM dd HHmm)', 'total_electron_energy_released (keV)', 'total_proton_energy_released (keV)','electron_energy_released_38_53 (particles/km2-s-ster-keV)','electron_energy_released_175_315 (particles/km2-s-ster-keV)','proton_energy_released_47_68 (particles/km2-s-ster-keV)','proton_energy_released_115_195 (particles/km2-s-ster-keV)','proton_energy_released_310_580 (particles/km2-s-ster-keV)','proton_energy_released_795_1193 (particles/km2-s-ster-keV)','proton_energy_released_1060_1900 (particles/km2-s-ster-keV)']]
selected_df1.columns = ['Time', 'energy_e', 'energy_p','energy_e_38_53','energy_e_175_315','energy_p_47_68','energy_p_115_195','energy_p_310_580','energy_p_795_1193','energy_p_1060_1090']
filename='D:\\_NILESH\\dissertation\\data\\cleaned\\'
selected_df1.to_csv(filename+'solar_flare_energy.csv')
selected_df1.head()
```
